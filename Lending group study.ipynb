```python
#importing panda, Numpy,matplotlib 
import numpy as np
import pandas as pd
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading data first 10 rows

data = pd.read_csv(r'D:\section\Upgrad\Lending study group\loan.csv')
data.head(10)
```

```python
# First look of overall data through describe function

data.describe()
```

```python
data.info() #column and row information
```

```python
#Dropping Columns having more than 60 Percent "empty" Values 

data.dropna(thresh=data.shape[0]*0.6,how='all',axis=1)

data.shape
```

```python
data.info()
```

# Univariate Analysis 

```python
# #Annual income plot
sns.boxplot(y=data["annual_inc"] )
plt.ylabel('annual_inc')
plt.show()

#plot show that most of applicants having income in range of 2 lac so more information we will have to remove outliers
```

```python
#After removing outliers
```

```python
outlier = data.annual_inc.quantile(.99)
print(outlier)

data = (data[data['annual_inc'] < outlier])

fig, axes = plt.subplots(1,2,figsize=(12,6))
sns.distplot(data['annual_inc'], hist=True, kde=True, color='g', ax=axes[0])

sns.boxplot(data['annual_inc'], ax=axes[1])
axes[0].set_title('Distribution of annual income', fontsize=15)
axes[1].set_title('Annual income range', fontsize=15)
plt.tight_layout()
plt.show()
```

From Above graph, We can draw followings conclusions:
1) Salary spreads seems to be in range of $40000 to $80000

2) Average salary of all applicants is below $60000

```python
# Analysis on loan status 
sns.countplot(x = 'loan_status', data = data)
plt.show()


```

```python
# Analysis on loan purpose
plt.figure(figsize=(13,5))
sns.countplot(x = 'purpose', data = data)
ax = sns.countplot(x="purpose", data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

# Graph show that most number of applicant from debt_consolidation which means they are taking loan to pay off existing loan.
```

```python
#Grade
sns.countplot(x='grade', data=data)
plt.show()
```

```python
# We will have dropp current segement since they have pay off loan
data = data[data['loan_status'] != 'Current']
```

# Segmented Univariate Analysis

```python
# Analysis on home ownership
data.groupby('loan_status')['home_ownership'].value_counts().unstack().plot(kind='bar', figsize=[15,8])
plt.show()
```

```python
# Analysis on loan term
data.groupby('loan_status')['term'].value_counts().unstack().plot(kind='bar', figsize=[10,6])
plt.show()
```

# Bivariate Analysis

```python
sns.barplot(y='annual_inc', x='pub_rec_bankruptcies', data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
#areawise destribution
g = data[data['loan_status']=='Charged Off'].groupby('addr_state')['loan_status'].count().reset_index()
plt.figure(figsize=(10,8))
sns.barplot(y='addr_state', x='loan_status', data=g)
plt.xlabel('Count of loan status to be defaulter')
plt.ylabel('State')
plt.show()
```
